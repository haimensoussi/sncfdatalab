```python
df = spark.table("sncf_trafic_dev.curated.train_status")
agg_df = df.groupBy("TrainType").agg(
    {"DelayMinutes": "avg", "TrainID": "count", "TravelDurationMinutes": "avg"}
).withColumnRenamed("avg(DelayMinutes)", "AvgDelayMinutes") \
 .withColumnRenamed("count(TrainID)", "TrainCount") \
 .withColumnRenamed("avg(TravelDurationMinutes)", "AvgTravelDurationMinutes")
display(agg_df)
```
